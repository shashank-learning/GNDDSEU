### Creating new database and collection in MongoDB

```python
import pymongo
```

```python
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
```

```python
tcs_db=myclient['tcs_db']
```

```python
emp_coll=tcs_db['emp_coll']
```

### However, the database and the collection are not created until you add some content to them. lets fill them with data.

```python
document = {
    "Name":"Aman",
    "Age":25,
    "State":"Gujarat",
    "Profile":"Developer",
    "Salary":55000,
    "Experience":3,
    "Skills":["PHP","MySQL"]   
   }
```

```python
emp_coll.insert_one(document)
```

### we have inserted data successfully. Now list all the database using list_database_names() method 

```python
for db in myclient.list_database_names():
    print(db)
```
